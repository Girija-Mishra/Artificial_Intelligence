```python
# Vacuum Cleaner Agent for Multiple Rooms
# 1 = Dirty, 0 = Clean

rooms = {
    "StoreRoom": [1, 1, 1],
    "Terrace":   [1, 1, 1],
    "Bedroom":   [1, 1, 1],
    "BR1":       [1, 1, 1],
    "Kitchen":   [1, 1, 1],
    "Hall":      [1, 1, 1]
}

room_order = ["StoreRoom", "Terrace", "Bedroom", "BR1", "Kitchen", "Hall"]

current_room_index = 0
current_position = 0  # 0 = Left, 1 = Center, 2 = Right

print("All rooms are dirty initially\n")

while current_room_index < len(room_order):
    room_name = room_order[current_room_index]
    room = rooms[room_name]

    print(f"Entering {room_name}")

    while current_position < 3:
        if room[current_position] == 1:
            print(f" Cleaning {room_name} - Position {current_position}")
            room[current_position] = 0
        current_position += 1

    print(f"{room_name} cleaned:", room, "\n")

    # Move to next room
    current_position = 0
    current_room_index += 1

# Check final environment
all_clean = all(
    rooms[room][pos] == 0
    for room in rooms
    for pos in range(3)
)

if all_clean:
    print("All rooms are cleaned successfully!")
    print("Performance: 100%")
    print("Thank you for using Mishra's797 vaccum cleaner")

print("\nFinal Environment State:")
for room, state in rooms.items():
    print(room, ":", state)
```
