```python
from collections import deque

# ---------- TREE STRUCTURE (ASCII) ----------
print("""
TREE STRUCTURE

                 a
            /    |    \\
           b     c     d
         /   \\          \\
        e     f          g
      /   \\            /   \\
     i     j           k     l
""")

# ---------- GRAPH ----------
graph = {
    'a': ['b', 'c', 'd'],
    'b': ['e', 'f'],
    'c': [],
    'd': ['g', 'h'],
    'e': ['i', 'j'],
    'f': [],
    'g': ['k', 'l'],
    'h': [],
    'i': [], 'j': [], 'k': [], 'l': []
}

# ---------- BFS FOR OPTIMAL PATH ----------
queue = deque(['a'])
visited = {'a'}
processed = []
parent = {'a': None}

step = 1
goal = 'l'

while queue:
    print(f"\nSTEP {step}")

    source = queue.popleft()
    print("Source Node      :", source)
    print("Adjacent Nodes   :", graph[source])

    for node in graph[source]:
        if node not in visited:
            visited.add(node)
            parent[node] = source
            queue.append(node)

    processed.append(source)

    print("Visited Nodes    :", visited)
    print("Processed Nodes  :", processed)

    if source == goal:
        break

    step += 1

# ---------- RECONSTRUCT OPTIMAL PATH ----------
path = []
current = goal
while current is not None:
    path.append(current)
    current = parent[current]

path.reverse()

print("\n✅ OPTIMAL PATH")
print("Path :", " → ".join(path))
```
