```python
LEFT = "LEFT"
RIGHT = "RIGHT"
SUCK = "SUCK"
NO_OP = "NO_OPERATION"

def vacuum_cleaner():
    location = 'A'
    status = {
        'A': 'Dirty',
        'B': 'Dirty'
    }

    print("Initial State:", location, status)

    while True:

        if status['A'] == 'Clean' and status['B'] == 'Clean':
            action = NO_OP
        elif status[location] == 'Dirty':
            action = SUCK
        elif location == 'A':
            action = RIGHT
        else:
            action = LEFT

        print("Action:", action)


        if action == SUCK:
            status[location] = 'Clean'
        elif action == RIGHT:
            location = 'B'
        elif action == LEFT:
            location = 'A'
        elif action == NO_OP:
            break

        print("State:", location, status)

    print("Final State:", location, status)



vacuum_cleaner()
```
